```python

for i in range(len(df)):
    description = df.loc[i, 'description']
    company = df.loc[i, 'company']
    name =  df.loc[i, 'first name']
    my_prompt = f'''I provide businesses with the opportunity to push & promote products through influencer marketing. We outreach to these leads via email and then our goal is to book a call and then convert them to a client and run influencer marketing campaigns for them. Create a cold email that can reach out to the following company: [{company}]. The recipient's name is [{name}]. My name is Cosette. Here is a description of the company: [{description}]. Make the cold email start off complimenting the recipient's business. Do not use any placeholders in this cold email. Use openers such as "I love that" or "I really enjoyed" But ONLY 50% of the time. Create some of your own openers as well'''
    try:
        response = openai.Completion.create(
            engine='text-davinci-003',
            prompt=my_prompt,
            max_tokens=500,
            n=1,
            stop=None,
            temperature=0.7)
        email_text = response["choices"][0]["text"]
        email_text = email_text.replace("\n", "")
        df.loc[i, 'fl'] = email_text
    except ValueError:
        print(ValueError)
    time.sleep(1)
    print(company, "cold email complete.")
```

```python
for i in range(len(df)):

    if df.loc[i, 'fl'][0]==".":
        df.loc[i, 'fl'] = df.loc[i, 'fl'][1:]
        
    if df.loc[i, 'fl'][0]=="...":
        df.loc[i, 'fl'] = df.loc[i, 'fl'][3:]
    print(df.loc[i, 'fl'],"\n")

df.head(20)

```
